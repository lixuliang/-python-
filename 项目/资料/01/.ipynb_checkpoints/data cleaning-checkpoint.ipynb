'''
【项目01】  商铺数据加载及存储

作业要求：
1、成功读取“商铺数据.csv”文件
2、解析数据，存成列表字典格式：[{'var1':value1,'var2':value2,'var3':values,...},...,{}]
3、数据清洗：
① comment，price两个字段清洗成数字
② 清除字段缺失的数据
③ commentlist拆分成三个字段，并且清洗成数字
4、结果存为.pkl文件

'''

```python
f = open(r'D:\user_profile\python\2\商铺数据.csv','r',encoding='utf8')
for i in f.readlines()[:10]:
    print(i.split(',')[4])
```

```python
# 创建comment、price、commentlist清洗函数

# commect 字段清洗,空格为分割符,选取结果list为第一个点评数,被转化为整型
def fcm(s):
    while '条' in s:
        return int(s.split(' ')[0])
    else:
        return('缺失数据')
    
# price 字段清洗,￥为分割符,选择结果list的最后一个为价格,并转化成浮点型
def fcp(s):
    while '￥' in s:
        return float(s.split('￥')[-1])
    else:
        return('缺失数据')

# commentlist 字段清洗,空格为分隔符,并把,质量,环境,服务转换为浮点型
def fcl(s):
    if ' ' in s:
        quality = float(s.split('                                ')[0][2:])
        environment = float(s.split('                                ')[1][2:])
        service = float(s.split('                                ')[2][2:-1])
        return [quality,environment,service]
    else:
        return('缺失数据')
        
```

```python
# 数据清洗

dataset = [] # 创建空列表,存放结果
f.seek(0)
for line in f.readlines()[1:]:
    data = line.split(',')
    classify = data[0]                # 提取分类
    name = data[1]                    # 提取名称
    comment = fcm(data[2])            # 提取评论数
    star = data[3]                    # 提取星级
    price = fcp(data[4])              # 提取价格
    address = data[5]                 # 提取地址
    quality = fcl(data[6])[0]         # 提取质量
    environment = fcl(data[6])[1]     # 提取环境
    service = fcl(data[6])[2]         # 提取服务
    
    if '缺失数据' not in [comment,price,quality]:
        data_re = [
                   ['classify',classify],
                   ['name',name],
                   ['comment',comment],
                   ['star',star],
                   ['price',price],
                   ['address',address],
                   ['quality',quality],
                   ['environment',environment],
                   ['service',service]]
        dataset.append(dict(data_re))
        
    else:
        continue

print(dataset)
```

```python
import pickle

# 把数据以pickle写到本地
f_w = open(r'D:\user_profile\python\project\1\shop.pkl','wb')
pickle.dump(dataset,f_w)
f_w.close()
print('finish')
```
